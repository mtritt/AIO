```python
# aivietnam - example

# change list while iterating it

# create a list of numbers
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    # print the num
    print(num)
    
    # check if the num is equal to 3
    if (num == 3):
        # add 3 to the list
        numbers.append(3)
```

```python
# aivietnam - copy in python

import copy

list1 = [1, 2, 3, 4, 5]
list2 = copy.copy(list1)

print('List 1: ', list1)
print('List1 address: ', id(list1))

print()
print('List 2: ', list2)
print('List2 address: ', id(list2))
```

```python
list1[3] = 9
print('List 1: ', list1)
print('List 2: ', list2)
```

```python
# aivietnam - copy in python

import copy

list1 = [[1, 2], 3]
list2 = copy.copy(list1)

print('List 1: ', list1)
print('List1 address: ', id(list1))

print()
print('List 2: ', list2)
print('List2 address: ', id(list2))
```

```python
print('List1[0] address: ', id(list1[0]))
print('List2[0] address: ', id(list2[0]))
```

```python
list1[0][0] = 9
print('List 1: ', list1)
print('List 2: ', list2)
```

```python
# aivietnam - copy in python

import copy

list1 = [[1, 2], 3]
list2 = copy.deepcopy(list1)

print('List 1: ', list1)
print('List1 address: ', id(list1))

print()
print('List 2: ', list2)
print('List2 address: ', id(list2))
```

```python
print('List1[0] address: ', id(list1[0]))
print('List2[0] address: ', id(list2[0]))
```

```python
list1[0][0] = 9
print('List 1: ', list1)
print('List 2: ', list2)
```
